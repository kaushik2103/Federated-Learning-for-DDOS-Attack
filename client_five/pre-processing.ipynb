```python
def remove_highly_correlated_features(df, target_column_name, threshold=0.85):
    if target_column_name not in df.columns:
        raise ValueError(f"Target column '{target_column_name}' not found in the DataFrame.")

    features_df = df.drop(columns=[target_column_name]).select_dtypes(include=[np.number])

    if features_df.shape[1] == 0:
        print("[INFO] No numeric features found. Returning original DataFrame.")
        return df

    corr_matrix = features_df.corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    to_drop = {column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)}

    if not to_drop:
        print("\n[INFO] No columns to remove. All feature correlations are below the threshold.")
        return df

    print(f"\n[INFO] Identified {len(to_drop)} columns to remove (threshold={threshold}):")
    print(sorted(list(to_drop)))
    df_cleaned = df.drop(columns=to_drop)
    print("\n[INFO] Successfully removed the columns.")

    return df_cleaned
```

```python
final_processed_df = remove_highly_correlated_features(df, 'Label', threshold=0.85)
```
